```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

pd.set_option('display.max_columns', None)
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
data = pd.read_csv('data/NBA_Rookies_1979_2020(greater than 2 seasons).csv')
print(data.shape)
data.head()
```

```python
team_map ={
    'NOK':'NOP',#pelicans
    'SDC':'LAC',# clippers
    #'NOP':'NOP',#pelicans
    'CHO':'CHH', #hornets
    'KCK':'SAC', # kings
    'VAN':'MEM', # grizzs
    'NOH':'NOP', # pelicans
    #'BRK':'BRK', # Nets
    'CHA':'CHH', # hornets
    #'OKC':'OKC', # okc
    'WAS':'WSH', #wizards
    #'TOR':'TOR', # raptors
    'WSB':'WSH', # wizards
    'SEA':'OKC', # sonics now okc
    'NJN':'BRK',#nets
    
}
```

```python
for old_team, new_team in team_map.items():
    data['Team'].replace(old_team, new_team, inplace=True)
```

```python
data.rename(columns = {'Target':'target>=2'}, inplace = True)
```

```python
data['target>=3'] = data.apply(lambda x : 1 if x.Career >=3  else 0 ,axis=1)
data['target>=4'] = data.apply(lambda x : 1 if x.Career >=4  else 0 ,axis=1)
data['target>=5'] = data.apply(lambda x : 1 if x.Career >=5  else 0 ,axis=1)
```

```python
train = data[data.Year<=2017]
```

```python
print(train.shape)
train.head()
```

```python
nums ='Age	Games	MP	PTS	FG	FGA	FG%	3P	3PA	3P%	FT	FTA	FT%	ORB	DRB	TRB	AST	STL	BLK	TOV	PF	MPpg	PTSpg	FGpg	3Ppg	FTpg	ORBpg	DRBpg	TRBpg	ASTpg	STLpg	BLKpg	TOVpg	PFpg'.split('	')
nums
```

```python
cats = 'Team	Year	Conf'.split('	')
cats
```

```python
predictors = nums +cats
```

```python
targets = 'target>=2	target>=3	target>=4	target>=5	Career'.split('	')
targets
```

```python
import pycaret
pycaret.__version__
```

```python
#For Classification
from pycaret.classification import *
```

```python
training_data = train[predictors+['target>=5']]
```

```python
exp_clf101 = setup(data = training_data, target = 'target>=5', session_id=123) 
```

```python
compare_models(fold = 10,sort = 'Recall')
```

```python
exp_clf102 = setup(data = training_data, target = 'target>=5', session_id=123,
                      multicollinearity_threshold= 0.8,
                    remove_multicollinearity = True,
                      polynomial_features= True,
                    polynomial_degree = 2) 
```

```python
compare_models(fold = 10,sort = 'Accuracy')
```

```python
exp_clf103 = setup(data = training_data, target = 'target>=5', session_id=123,
                      multicollinearity_threshold= 0.8,
                    remove_multicollinearity = True,
                      polynomial_features= True,
                    polynomial_degree = 2,
                      create_clusters= True,
                    cluster_iter=10) 
```
