```python
for idx in range(len(players)):
    if idx == 0:
        continue
    player_rows = players[idx].find_all("td")
    a=1
    for element in player_rows:
        #print(element.text.strip())
        #print(',',a)
        if a == 1:
            #print(element.text.strip())
            index_.append(element.text.strip())
        if a == 2:
            #print(element.text.strip())
            player_.append(element.text.strip())
        if a == 3:
            #print(element.text.strip())
            position_.append(element.text.strip())
        if a == 4:
            #print(element.text.strip())
            team_.append(element.text.strip())
        a+=1
```

```python
def relevant_cols(df):
    df=df.copy()
    num_cols='Games	MP	FG	FGA	3P	3PA	FT	FTA	ORB	TRB	AST	STL	BLK	TOV	PF	PTS	FG%	3P%	FT%	MPpg	PTSpg	TRBpg	ASTpg'.split('	')
    for col in num_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df['FGpg'] = df['FG']/df['Games'] #'FGpg':'field_goals_attempted',
    df['3Ppg'] = df['3P']/df['Games'] #'3Ppg':'three_points_attempted',
    df['FTpg'] = df['FT']/df['Games']#'FTpg':'field_goals_attempted',
    df['ORBpg'] = df['ORB']/df['Games'] #'ORBpg':'offensive_rebounds',
    for col in df:
        df[col].replace(np.inf,0, inplace=True)
        df[col].replace(np.nan,0, inplace=True)
    df['DRBpg'] = df['TRBpg']-df['ORBpg'] #'DRBpg':'defensive_rebounds',
    df['STLpg'] = df['STL']/df['Games']  #'STLpg':'steals',
    df['BLKpg'] = df['BLK']/df['Games'] #'BLKpg':'blocks',
    df['TOVpg'] = df['TOV']/df['Games'] #'TOVpg':'turnovers',
    df['PFpg'] = df['PF']/df['Games'] #'PFpg':,
    for col in df:
        df[col].replace(np.inf,0, inplace=True)
        df[col].replace(np.nan,0, inplace=True)
    return df
```

```python
data_rook = relevant_cols(data_rook)
```
